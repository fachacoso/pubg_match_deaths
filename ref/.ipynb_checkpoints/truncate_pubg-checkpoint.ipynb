# Truncate PUBG #

This is a notebook to truncate the original PUBG dataset from Kaggle to something slightly easier to workwith. This picks data from all matches on 12/10/2017. 

Full dataset: https://www.kaggle.com/skihikingkevin/pubg-match-deaths

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### `Aggregate` Dataset

```python
agg_all = pd.DataFrame(columns=['date', 'game_size', 'match_id', 'match_mode', 'party_size',
       'player_assists', 'player_dbno', 'player_dist_ride', 'player_dist_walk',
       'player_dmg', 'player_kills', 'player_name', 'player_survive_time',
       'team_id', 'team_placement'])
for i in np.arange(5):
    agg_all = agg_all.append(pd.read_csv('aggregate/agg_match_stats_%d.csv' %i))

agg_all['date'] = pd.to_datetime(agg_all['date'])
agg_all.reset_index(drop=True, inplace=True)
agg_all.head(10)
```

```python
agg_short = agg_all[(agg_all['date'].dt.year == 2017) 
                    & (agg_all['date'].dt.month == 12)
                    & (agg_all['date'].dt.day == 10)
                   ]
agg_short.reset_index(drop=True, inplace=True)
```

```python
agg_short.to_csv('aggregate.csv')
```

```python
games = agg_short['match_id'].unique()
```

### `Deaths` Dataset

```python
deaths_all = pd.DataFrame(columns=['killed_by', 'killer_name', 'killer_placement', 'killer_position_x',
       'killer_position_y', 'map', 'match_id', 'time', 'victim_name',
       'victim_placement', 'victim_position_x', 'victim_position_y'])
for i in np.arange(5):
    deaths_all = deaths_all.append(pd.read_csv('deaths/kill_match_stats_final_%d.csv' %i))

deaths_all.reset_index(drop=True, inplace=True)
deaths_all.head(10)
```

```python
deaths_short = deaths_all[deaths_all['match_id'].isin(games)]
deaths_short.reset_index(drop=True, inplace=True)
```

```python
deaths_short.to_csv('deaths.csv')
```
