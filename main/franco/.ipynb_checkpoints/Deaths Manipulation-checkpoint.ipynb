```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.image as mpimg 
map_img = mpimg.imread('img/erangel.jpg') 
%matplotlib inline

deaths = pd.read_csv('original data/deaths.csv')
deaths = deaths[deaths['victim_position_x'] != 0.0] #deaths without Early Leaving 
ids = deaths["match_id"].unique() #Converts deaths match_id's to indexes
indices = np.arange(ids.size)
id_table = pd.DataFrame({
    'id' : ids,
    'game_id' : indices
})
deaths = deaths.merge(id_table, left_on="match_id", right_on="id").drop(columns=['match_id', 'id'])

deaths_e = deaths[deaths['map'] == 'ERANGEL'] #deaths in the map ERANGEL 
deaths_m = deaths[deaths['map'] == 'MIRAMAR'] #deaths in the map MIRAMAR #NOTE: MIRAMAR map not in game when truncated

#deaths by bluezone
bzdeaths = deaths[deaths['killed_by'] == 'Bluezone']
```
