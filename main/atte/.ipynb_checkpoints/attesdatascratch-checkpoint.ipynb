```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
aggregate_data = pd.read_csv("aggregate.csv")
deaths_data = pd.read_csv("deaths.csv")
```

### Universal Match Info Calculator

```python
match_info = {} # a dictionary with match_id as the keynd data about that game as the value

for row in aggregate_data.iterrows():
    row = row[1]
    match_id = row["match_id"][-10:]
    if (match_id in match_info):
        continue
    else:
        game_size = row["game_size"]
        party_size = row["party_size"]
        match_info[match_id] = {"game_size": game_size, "party_size": party_size}
```

### Bluezone/Total Count Calculator

```python
NUM_BOXES = 200
DIVISOR = 800000 // DENSITY

bluezone_count = [0 for _ in range(5)] # ind 1: solo, ind 2: duo, ind 4: quads
total_count = [0 for _ in range(5)] # ind 1: solo, ind 2: duo, ind 4: quads
coordinates_bluezone = [[0 for _ in range(NUM_BOXES)] for _ in range(NUM_BOXES)]

for row in deaths_data.iterrows():
    row = row[1]
    killed_by = row["killed_by"]
    count_index = match_info[row["match_id"][-10:]]["party_size"]
    if (killed_by == ""):
        continue
    total_count[count_index] += 1
    if (killed_by == "Bluezone"):
        bluezone_count[count_index] += 1
        x_box = int(row["victim_position_x"] // DIVISOR)
        y_box = int(row["victim_position_y"] // DIVISOR)
        x_box = min(NUM_BOXES-1, x_box)
        y_box = min(NUM_BOXES-1, y_box)
        coordinates_bluezone[y_box][x_box] += 1
```

```python
print("Solos: " + str((bluezone_count[1]/total_count[1])*100) + "%")
print("Duos: " + str((bluezone_count[2]/total_count[2])*100) + "%")
print("Quads: " + str((bluezone_count[4]/total_count[4])*100) + "%")
```

```python
bluezone_df = pd.DataFrame(coordinates_bluezone)
sns.heatmap(bluezone_df, robust=True)
```

```python
guns_dict = {"Down and Out": "DO",
             "P18C": "PT",
             "Bluezone": "BZ",
             "UMP9": "SMG",
             "S1897": "SG",
             "Punch": "ML",
             "M16A4": "AR",
             "Groza": "AR",
             "Mini 14": "DMR",
             "Tommy Gun": "SMG",
             "M416": "AR",
             "SCAR-L": "AR",
             "AKM": "AR",
             "S686": "SG",
             "S12K": "SG",
             "SKS": "DMR",
             "Micro UZI": "SMG",
             "Falling": "FL",
             "Kar98k": "SN",
             "Buggy": "VH",
             "Grenade": "EX",
             "R1895": "PT",
             "Vector": "SMG",
             "Uaz": "VH",
             "P1911": "PT",
             "Motorbike (SideCar)": "VH",
             "P92": "PT",
             "Hit by Car": "VH",
             "Drown": "DR",
             "RedZone": "RZ",
             "VSS": "SN",
             "Machete": "ML",
             "Mk14": "DMR",
             "Crossbow": "BOW",
             "Dacia": "VH",
             "Sickle": "ML",
             "M249": "LMG",
             "M24": "SN",
             "Pan": "ML",
             "death.Buff_FireDOT_C": "EX",
             "Motorbike": "VH",
             "Boat": "VH",
             "death.ProjMolotov_DamageField_C": "EX",
             "AWM": "SN",
             "Crowbar": "ML",
             "death.ProjMolotov_C": "EX"}

gun_cats = {'DR': "Drowning",
            'DO': "Down and Out",
            'EX': "Explosion",
            'VH': "Vehicle",
            'PT': "Pistol",
            'BOW': "Crossbow",
            'BZ': "Bluezone",
            'SN': "Sniper",
            'LMG': "Light Machine Gun",
            'ML': "Melee",
            'DMR': "Designated Marksman Rifle",
            'AR': "Assault Rifle",
            'FL': "Falling",
            'SG': "Shotgun",
            'SMG': "Submachine Gun",
            'RZ': "Red Zone"}
```

```python
gun_results = []

for row in deaths_data.iterrows():
    row = row[1]
    killer_x = row["killer_position_x"]
    killer_y = row["killer_position_y"]
    victim_x = row["victim_position_x"]
    victim_y = row["victim_position_y"]
    killed_by = guns_dict[row["killed_by"]]
    game_size = match_info[row["match_id"][-10:]]["game_size"]
    party_size = match_info[row["match_id"][-10:]]["party_size"]
    if (math.isnan(killer_x) or math.isnan(killer_y)):
        killer_x, killer_y = 0, 0
    gun_results.append([killer_x, killer_y, victim_x, victim_y, killed_by, game_size, party_size])
```

```python
uniqueGuns = set(guns_dict.values())
print(uniqueGuns)
```

```python
NUM_BOXES = 200
DIVISOR = 800000 / NUM_BOXES

counter = 0

rounded_gun_results_killer = [[{val: 0 for val in uniqueGuns} for _ in range(NUM_BOXES)] for _ in range(NUM_BOXES)]

for kill in gun_results:
    x_box = int(kill[0] // DIVISOR)
    y_box = int(kill[1] // DIVISOR)
    gun_cat = kill[4]
    x_box = min(x_box, NUM_BOXES-1)
    y_box = min(y_box, NUM_BOXES-1)
    rounded_gun_results_killer[y_box][x_box][gun_cat] += 1
```

```python
mapped_by_guns = {gun_cat: [[0 for _ in range(NUM_BOXES)] for _ in range(NUM_BOXES)] for gun_cat in uniqueGuns}

for i, gun_cat in enumerate(uniqueGuns):
    for y, box_y in enumerate(rounded_gun_results_killer):
        for x, box_x in enumerate(box_y):
            mapped_by_guns[gun_cat][y][x] = box_x[gun_cat]
```

```python
dataframes = []
gun_cat_inorder = []

for gun in uniqueGuns:
    gun_cat_inorder.append(gun)
    dataframes.append(pd.DataFrame(mapped_by_guns[gun]))
```

```python
cmap = sns.choose_light_palette(input="rgb")
```

# Maps

```python
cmap[0] = [1, 1, 1, 1]
plt.rcParams["figure.figsize"] = [16,9]
for i, dataframe in enumerate(dataframes):
    plt.figure(i)
    plt.title(gun_cats[gun_cat_inorder[i]] + " (" + str(i) + ")")
    if i in {0, 5, 8, 15}:
        sns.heatmap(dataframe, cmap=cmap, vmin=0, vmax=2)
    else:
        sns.heatmap(dataframe, cmap=cmap, robust=True)
```
